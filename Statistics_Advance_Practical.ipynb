```python
#3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(p, x):
    if x not in [0, 1]:
        return 0
    return p if x == 1 else 1 - p


p_success = 0.5
for x in [0, 1]:
    print(f'P(X={x}) =', bernoulli_pdf(p_success, x))
```

```python
# 4 Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram>
import numpy as np
import matplotlib.pyplot as plt
n = 10
p = 0.5
size = 10000


binomial_samples = np.random.binomial(n, p, size)

plt.figure(figsize=(8, 5))
plt.hist(binomial_samples, bins=np.arange(n+2)-0.5, density=True, alpha=0.7, color='blue', edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xticks(range(n+1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#6> Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution
import numpy as np
import matplotlib.pyplot as plt


low = 1
high = 10
size = 10000
uniform_samples = np.random.randint(low, high+1, size)


values, counts = np.unique(uniform_samples, return_counts=True)
cdf = np.cumsum(counts) / size
plt.figure(figsize=(8, 5))
plt.step(values, cdf, where='mid', color='blue', label='CDF')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title(f'Discrete Uniform Distribution CDF ({low} to {high})')
plt.xticks(range(low, high+1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
#9  Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

mean = 0
std_dev = 1
size = 10000


normal_samples = np.random.normal(mean, std_dev, size)


def calculate_z_scores(data):
    return (data - np.mean(data)) / np.std(data)


z_scores = calculate_z_scores(normal_samples)

plt.figure(figsize=(8, 5))
plt.hist(z_scores, bins=50, density=True, alpha=0.7, color='blue', edgecolor='black')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.title('Z-score Distribution')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#16  Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)5
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


def plot_standard_normal():
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.plot(x, y, color='blue', label='Standard Normal Curve')
    plt.xlabel('X')
    plt.ylabel('Density')
    plt.title('Standard Normal Distribution')
    plt.legend()
    plt.grid()
    plt.show()

sample_size = 30
num_samples = 1000
size = 10000


normal_samples = np.random.normal(5, 2, size)


sample_means = [np.mean(np.random.choice(normal_samples, sample_size)) for _ in range(num_samples)]


plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black', density=True)
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.title('Central Limit Theorem Demonstration')
plt.grid()
plt.show()


plot_standard_normal()
```

```python
#18 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
from scipy.stats import norm

def calculate_z_score(value, mean, std_dev):
    z_score = (value - mean) / std_dev
    probability = norm.cdf(z_score)
    return z_score, probability

# Example values
data_point = 75
mean = 60
std_dev = 10

z_score, prob = calculate_z_score(data_point, mean, std_dev)

print(f"Z-score: {z_score:.2f}")
print(f"Probability of being less than {data_point}: {prob:.4f}")
```

```python
#19 Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

np.random.seed(42)
data = np.random.normal(50, 10, 30)

pop_mean = 50
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)
z_score = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f"Z-score: {z_score:.2f}, P-value: {p_value:.4f}")

```

```python
#21 Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
import scipy.stats as stats

np.random.seed(42)
data = np.random.normal(50, 10, 1000)

mean, std_err = np.mean(data), stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=std_err)

print(f"95% Confidence Interval: {ci}")

```

```python
#24  Write a Python program to compare the standard deviations of two datasets and visualize the difference
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
data1 = np.random.normal(50, 10, 1000)
data2 = np.random.normal(50, 20, 1000)


std1 = np.std(data1, ddof=1)
std2 = np.std(data2, ddof=1)

plt.hist(data1, bins=30, alpha=0.5, label=f'Dataset 1 (={std1:.2f})', color='blue', edgecolor='black')
plt.hist(data2, bins=30, alpha=0.5, label=f'Dataset 2 (={std2:.2f})', color='red', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Comparison of Standard Deviations')
plt.legend()
plt.grid()
plt.show()

print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")
```

```python
#25  Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
import numpy as np


np.random.seed(42)
data = np.random.normal(50, 10, 1000)

range_val = np.ptp(data)
iqr_val = np.percentile(data, 75) - np.percentile(data, 25)


print(f"Range: {range_val:.2f}")
print(f"Interquartile Range (IQR): {iqr_val:.2f}")
```

```python
#26  Implement Z-score normalization on a dataset and visualize its transformation
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
data = np.random.normal(50, 10, 1000)


mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std


plt.hist(data, bins=30, alpha=0.5, label='Original', color='blue')
plt.hist(z_scores, bins=30, alpha=0.5, label='Z-score Normalized', color='red')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Z-score Normalization')
plt.legend()
plt.grid()
plt.show()


print(f"Original Mean: {mean:.2f}, Standard Deviation: {std:.2f}")
```
